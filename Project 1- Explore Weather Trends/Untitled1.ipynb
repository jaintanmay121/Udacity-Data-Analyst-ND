```python
import pandas as pd
df=pd.read_csv('results.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.year,df.cat)
plt.plot(df.year,df.gat)
plt.legend(['New Delhi', 'Global'])
plt.xlabel('Year')
plt.ylabel('Temprature ( °C)')
plt.show()
```

```python
plt.plot(df.year,df.cat.rolling(window=7).mean())
plt.plot(df.year,df.gat.rolling(window=7).mean())
plt.title('7 Year Moving Average Weather Trends')
plt.legend(['New Delhi', 'Global'])
plt.xlabel('Year')
plt.ylabel('Temprature ( °C)')
plt.show()
```
